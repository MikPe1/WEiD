```python
# 1. Utworzyć wykres liniowy funkcji 10x + 5 dla dziedziny w zakresie [-10, 60]. Serię danych zwizualizować niebieskimi kwadratami.
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 60, 100)
plt.title("Wykres funkcji liniowej y = 10*x + 5")
y = 10*x + 5
plt.plot(x, y, 'bs')  # 'bs' oznacza niebieskie kwadraty
plt.show()
```

```python
# 2. Utworzyć wykres punktowo-liniowy funkcji liniowej 20x - 4 dla dziedziny w zakresie [0, 72]. Serię danych oznaczyć kolorem czerwonym. Wykres zatytułować "funkcja liniowa". Oznaczyć etykiety osi x oraz y.
x = np.linspace(0, 72, 100)
y = 20*x - 4
plt.plot(x, y, 'ro-', label='funkcja y=20*x - 4')
plt.title("Wykres punktowo-liniowy funkcji liniowej")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 3. Utworzyć wykresy liniowe funkcji liniowych o liczbie wyznaczonej pseudolosowo dla dziedziny w zakresie wyznaczonym pseudolosowo. Współczynniki a oraz b dobrać pseudolosowo. Na wykresie umieścić legendę definiującą każdą z serii danych.
import random

x = np.linspace(random.randint(-100, 0), random.randint(0, 100), 200)
for i in range(random.randint(2, 5)):  # Losowa liczba funkcji
    a = random.uniform(-10, 10)
    b = random.uniform(-50, 50)
    y = a*x + b
    plt.plot(x, y, label=f'y = {a:.2f}x + {b:.2f}')
plt.legend()
plt.show()
```

```python
# 4. Utworzyć siatkę wykresów dla wykresów funkcji liniowych o liczbie wyznaczonej pseudolosowo dla dziedziny w zakresie wyznaczonym pseudolosowo. Współczynniki a oraz b dobrać pseudolosowo. Na wykresie umieścić legendę definiującą każdą z serii danych.
fig, axs = plt.subplots(2, 2)  # Tworzy siatkę 2x2
for ax in axs.flatten():
    x = np.linspace(random.randint(-100, 0), random.randint(0, 100), 200)
    a = random.uniform(-10, 10)
    b = random.uniform(-50, 50)
    y = a*x + b
    ax.plot(x, y, label=f'y = {a:.2f}x + {b:.2f}')
    ax.legend()
plt.show()
```

```python
# 5. Utworzyć wykres słupkowy przedstawiający liczbę sprzedanych wycieczek do Sri Lanki, Malediwów, wysp Galapagos, Grecji, wyspy Bali, Hiszpanii i Włoch. Liczbę sprzedanych wycieczek wyznaczyć pseudolosowo. Nadać dowolny tytuł wykresu.
destinations = ["Sri Lanka", "Malediwy", "Galapagos", "Grecja", "Bali", "Hiszpania", "Włochy"]
sales = [random.randint(50, 200) for _ in destinations]
plt.bar(destinations, sales)
plt.title("Liczba sprzedanych wycieczek")
plt.show()
```

```python
# 6. Utworzyć wykres kołowy liczby projektów według języka programowania. Liczbę projektów i języki dopasować pseudolosowo. Wygenerować legendę.
languages = ["Python","R", "Java", "C#", "JavaScript"]
projects = [random.randint(1, 100) for _ in languages]
plt.pie(projects, labels=languages, autopct='%1.1f%%')
plt.legend()
plt.show()
```

```python
# 7. Utworzyć siatkę 2 wykresów kołowych:
# - pierwszy wykres będzie zawierał udział czterech wybranych marek samochodów w rynku. Liczebności wygenerować pseudolosowo
# - drugi wykres będzie zawierał udział czterech wybranych producentów laptopów w rynku. Liczebności wygenerować pseudolosowo.
# Dla każdego wykresu wygenerować legendę oraz nadać tytuł.

brands = ["BMW", "Audi", "Mercedes", "Toyota"]
computers = ["Lenovo", "HP", "Dell", "Apple"]
car_counts = [random.randint(100, 1000) for _ in brands]
computer_counts = [random.randint(100, 1000) for _ in computers]

fig, axs = plt.subplots(1, 2)
axs[0].pie(car_counts, labels=brands, autopct='%1.1f%%')
axs[0].set_title("Udział marek samochodów")
axs[1].pie(computer_counts, labels=computers, autopct='%1.1f%%')
axs[1].set_title("Udział producentów laptopów")

for ax in axs:
    ax.legend()
plt.show()
```

```python
#8. Utworzyć siatkę pseudolosowych wykresów punktowych oraz ułożyć je "w choinkę". Wygenerować "choinkę" wykresów o pseudolowej wysokości.
import matplotlib.pyplot as plt
import numpy as np

def generate_random_scatter(ax):
    x = np.random.rand(100)
    y = np.random.rand(100)
    ax.scatter(x, y)
    ax.axis('off')  # Usunięcie osi dla czytelności

# Ustalenie wysokości choinki
wysokosc_choinki = random.randint(1,10)

# Tworzenie figury
fig, axs = plt.subplots(wysokosc_choinki, wysokosc_choinki*2-1, figsize=(15, 10))

# Ustawienie tła
fig.patch.set_facecolor('whitesmoke')

# Iteracja przez siatkę subplotów i tworzenie wykresów w kształcie choinki
for i in range(wysokosc_choinki):
    for j in range(wysokosc_choinki*2-1):
        # Sprawdzenie, czy dany subplot powinien zostać włączony (jest częścią "choinki")
        if wysokosc_choinki - 1 - i <= j <= wysokosc_choinki - 1 + i:
            generate_random_scatter(axs[i, j])
        else:
            axs[i, j].axis('off')  # Wyłączenie subplotów poza "choinką"

# Usunięcie niepotrzebnej przestrzeni między wykresami
plt.tight_layout()

# Wyświetlenie choinki z wykresów
plt.show()
```
