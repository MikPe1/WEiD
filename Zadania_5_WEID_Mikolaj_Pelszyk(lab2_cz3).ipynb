```python
# 1. Przygotować liniowy wykres 3D, w którym znajdą się wartości z zakresu współrzędnych wyznaczonego pseudolosowo oraz liczebności dobranej pseudolosowo. Ponadto wartości na osiach X, Y, Z zostaną przekształcone przez funkcje trygonometryczne (dostępne w bibliotece NumPy) wyznaczone pseudolosowo. Przykładowo, wartości na osi X zostaną przekształcone funkcją sinus, na osi Y funkcją cosinus, a na osi Z funkcją tangens. W ramach jednej wizualizacji funkcje nie mogą się powtarzać.

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generowanie pseudolosowych danych
rng = np.random.default_rng()
x = rng.random(100)
y = rng.random(100)
z = rng.random(100)

function = np.random.choice([np.sin, np.cos, np.tan])


# Przekształcenie danych przy użyciu funkcji trygonometrycznych
x_transformed = function(x)
y_transformed = function(y)
z_transformed = function(z)

# Rysowanie wykresu
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x_transformed, y_transformed, z_transformed, color = 'lightblue')
ax.set_xlabel('Sin(X)')
ax.set_ylabel('Cos(Y)')
ax.set_zlabel('Tan(Z)')

plt.show()
```

```python
# 2. Przygotować punktowy wykres 3D, gdzie znajdą się punkty z zakresu współrzędnych wyznaczonego pseudolosowo oraz liczebności dobranej pseudolosowo. Każdemu punktowi nadać pseudolosowo dobrany kształt (kwadrat, trójkąt, koło) oraz pseudolosowy kolor (zielony, czerwony, niebieski).

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Generowanie pseudolosowych danych
rng = np.random.default_rng()
n = 100
x = rng.random(n) * 10
y = rng.random(n) * 10
z = rng.random(n) * 10
shapes = ['o', '^', 's']  # koło, trójkąt, kwadrat
colors = ['g', 'r', 'b']  # zielony, czerwony, niebieski

# Rysowanie wykresu
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for i in range(n):
    ax.scatter(x[i], y[i], z[i], marker=shapes[rng.integers(len(shapes))], color=colors[rng.integers(len(colors))])

plt.show()

```

```python
# 3. Przygotować wykres powierzchniowy 3D, którego podstawą będzie siatka o zakresie dobranym pseudolosowo. Współrzędnymi powierzchni na osi Z będzie odległość między współrzędnymi znajdującymi się na siatce przetransformowane funkcją trygonometryczną dobraną pseudolosowo.

import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import random

# Ustawienia początkowe dla zakresu siatki
x_start, x_end = random.randint(-10, 0), random.randint(0, 10)
y_start, y_end = random.randint(-10, 0), random.randint(0, 10)

# Tworzenie siatki
x = np.linspace(x_start, x_end, 50)
y = np.linspace(y_start, y_end, 50)
x, y = np.meshgrid(x, y)

# Wybór funkcji trygonometrycznej
trig_functions = [np.sin, np.cos, np.tan]
function = random.choice(trig_functions)

# Obliczanie wartości na osi Z
# Zastosowanie funkcji trygonometrycznej do obliczenia odległości
z = function(np.sqrt(x**2 + y**2))

# Tworzenie wykresu
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Tworzenie wykresu powierzchniowego
surf = ax.plot_surface(x, y, z, cmap='viridis')

# Dodanie paska kolorów dla głębi
fig.colorbar(surf)

# Ustawienia etykiet osi
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis (Distance)')

# Wyświetlanie wykresu
plt.show()
```

```python
# 4. Przygotować "choinkę" o wysokości dobranej pseudolosowo, która będzie zawierała pseudolosowe wykresy punktowe z zadania 2.
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import random

rng = np.random.default_rng()
n = 100
x = rng.random(n) * 10
y = rng.random(n) * 10
z = rng.random(n) * 10
shapes = ['o', '^', 's']
colors = ['g', 'r', 'b']

# Funkcja do generowania pseudolosowych wykresów punktowych w przestrzeni 3D
def generate_random_scatter(ax):
    for _ in range(n):
        ax.scatter(rng.random() * 10, rng.random() * 10, rng.random() * 10,
                   marker=shapes[rng.integers(len(shapes))], color=colors[rng.integers(len(colors))])

# Ustalenie wysokości choinki
wysokosc_choinki = random.randint(3, 7)

# Tworzenie figury
fig = plt.figure()

# Iteracja przez siatkę subplotów i tworzenie wykresów w kształcie choinki
for i in range(wysokosc_choinki):
    for j in range(wysokosc_choinki*2-1):
        if wysokosc_choinki - 1 - i <= j <= wysokosc_choinki - 1 + i:
            ax = fig.add_subplot(wysokosc_choinki, wysokosc_choinki*2-1, i*(wysokosc_choinki*2-1)+j+1, projection='3d')
            generate_random_scatter(ax)
        else:
            fig.add_subplot(wysokosc_choinki, wysokosc_choinki*2-1, i*(wysokosc_choinki*2-1)+j+1).axis('off')  # Wyłączenie subplotów poza "choinką"

# Ustawienie tła
fig.patch.set_facecolor('whitesmoke')

# Usunięcie niepotrzebnej przestrzeni między wykresami
plt.tight_layout()

# Wyświetlenie choinki z wykresów punktowych
plt.show()
```
