```python
# 1. Przygotowa liniowy wykres 3D, w ktrym znajd si wartoci z zakresu wsprzdnych wyznaczonego pseudolosowo oraz liczebnoci dobranej pseudolosowo. Ponadto wartoci na osiach X, Y, Z zostan przeksztacone przez funkcje trygonometryczne (dostpne w bibliotece NumPy) wyznaczone pseudolosowo. Przykadowo, wartoci na osi X zostan przeksztacone funkcj sinus, na osi Y funkcj cosinus, a na osi Z funkcj tangens. W ramach jednej wizualizacji funkcje nie mog si powtarza.

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generowanie pseudolosowych danych
rng = np.random.default_rng()
x = rng.random(100)
y = rng.random(100)
z = rng.random(100)

function = np.random.choice([np.sin, np.cos, np.tan])


# Przeksztacenie danych przy uyciu funkcji trygonometrycznych
x_transformed = function(x)
y_transformed = function(y)
z_transformed = function(z)

# Rysowanie wykresu
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x_transformed, y_transformed, z_transformed, color = 'lightblue')
ax.set_xlabel('Sin(X)')
ax.set_ylabel('Cos(Y)')
ax.set_zlabel('Tan(Z)')

plt.show()
```

```python
# 2. Przygotowa punktowy wykres 3D, gdzie znajd si punkty z zakresu wsprzdnych wyznaczonego pseudolosowo oraz liczebnoci dobranej pseudolosowo. Kademu punktowi nada pseudolosowo dobrany ksztat (kwadrat, trjkt, koo) oraz pseudolosowy kolor (zielony, czerwony, niebieski).

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Generowanie pseudolosowych danych
rng = np.random.default_rng()
n = 100
x = rng.random(n) * 10
y = rng.random(n) * 10
z = rng.random(n) * 10
shapes = ['o', '^', 's']  # koo, trjkt, kwadrat
colors = ['g', 'r', 'b']  # zielony, czerwony, niebieski

# Rysowanie wykresu
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for i in range(n):
    ax.scatter(x[i], y[i], z[i], marker=shapes[rng.integers(len(shapes))], color=colors[rng.integers(len(colors))])

plt.show()

```

```python
# 3. Przygotowa wykres powierzchniowy 3D, ktrego podstaw bdzie siatka o zakresie dobranym pseudolosowo. Wsprzdnymi powierzchni na osi Z bdzie odlego midzy wsprzdnymi znajdujcymi si na siatce przetransformowane funkcj trygonometryczn dobran pseudolosowo.

import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import random

# Ustawienia pocztkowe dla zakresu siatki
x_start, x_end = random.randint(-10, 0), random.randint(0, 10)
y_start, y_end = random.randint(-10, 0), random.randint(0, 10)

# Tworzenie siatki
x = np.linspace(x_start, x_end, 50)
y = np.linspace(y_start, y_end, 50)
x, y = np.meshgrid(x, y)

# Wybr funkcji trygonometrycznej
trig_functions = [np.sin, np.cos, np.tan]
function = random.choice(trig_functions)

# Obliczanie wartoci na osi Z
# Zastosowanie funkcji trygonometrycznej do obliczenia odlegoci
z = function(np.sqrt(x**2 + y**2))

# Tworzenie wykresu
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Tworzenie wykresu powierzchniowego
surf = ax.plot_surface(x, y, z, cmap='viridis')

# Dodanie paska kolorw dla gbi
fig.colorbar(surf)

# Ustawienia etykiet osi
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis (Distance)')

# Wywietlanie wykresu
plt.show()
```

```python
# 4. Przygotowa "choink" o wysokoci dobranej pseudolosowo, ktra bdzie zawieraa pseudolosowe wykresy punktowe z zadania 2.
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import random

rng = np.random.default_rng()
n = 100
x = rng.random(n) * 10
y = rng.random(n) * 10
z = rng.random(n) * 10
shapes = ['o', '^', 's']
colors = ['g', 'r', 'b']

# Funkcja do generowania pseudolosowych wykresw punktowych w przestrzeni 3D
def generate_random_scatter(ax):
    for _ in range(n):
        ax.scatter(rng.random() * 10, rng.random() * 10, rng.random() * 10,
                   marker=shapes[rng.integers(len(shapes))], color=colors[rng.integers(len(colors))])

# Ustalenie wysokoci choinki
wysokosc_choinki = random.randint(3, 7)

# Tworzenie figury
fig = plt.figure()

# Iteracja przez siatk subplotw i tworzenie wykresw w ksztacie choinki
for i in range(wysokosc_choinki):
    for j in range(wysokosc_choinki*2-1):
        if wysokosc_choinki - 1 - i <= j <= wysokosc_choinki - 1 + i:
            ax = fig.add_subplot(wysokosc_choinki, wysokosc_choinki*2-1, i*(wysokosc_choinki*2-1)+j+1, projection='3d')
            generate_random_scatter(ax)
        else:
            fig.add_subplot(wysokosc_choinki, wysokosc_choinki*2-1, i*(wysokosc_choinki*2-1)+j+1).axis('off')  # Wyczenie subplotw poza "choink"

# Ustawienie ta
fig.patch.set_facecolor('whitesmoke')

# Usunicie niepotrzebnej przestrzeni midzy wykresami
plt.tight_layout()

# Wywietlenie choinki z wykresw punktowych
plt.show()
```
