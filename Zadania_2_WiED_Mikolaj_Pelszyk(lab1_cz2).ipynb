```python
import pandas as pd
import numpy as np
# 1. Wczytać zbiór danych z pliku movie_new.csv do ramki movie_new. Przy użyciu parametru names funkcji read_csv nadać następujące nazwy kolumnom: color,director_name,num_critic_for_reviews,duration,director_facebook_likes,actor_3_facebook_likes,actor_2_name,actor_1_facebook_likes,gross,genres,actor_1_name,movie_title,num_voted_users,cast_total_facebook_likes,actor_3_name,facenumber_in_poster,plot_keywords,movie_imdb_link,num_user_for_reviews,language,country,content_rating,budget,title_year,actor_2_facebook_likes,imdb_score,aspect_ratio,movie_facebook_likes.
column_names = ['color', 'director_name', 'num_critic_for_reviews', 'duration', 'director_facebook_likes', 'actor_3_facebook_likes'
                , 'actor_2_name', 'actor_1_facebook_likes', 'gross', 'genres', 'actor_1_name', 'movie_title', 'num_voted_users', 'cast_total_facebook_likes'
                , 'actor_3_name', 'facenumber_in_poster', 'plot_keywords', 'movie_imdb_link', 'num_user_for_reviews', 'language', 'country', 'content_rating'
                , 'budget', 'title_year', 'actor_2_facebook_likes', 'imdb_score', 'aspect_ratio', 'movie_facebook_likes']

movie_new = pd.read_csv('../WIED/movie_new.csv', names=column_names, sep=',', header=0)
movie_new.head()
```

```python
# 2. Utworzyć ramkę movie_num, w której będą zawarte jedynie kolumny z wartościami typu numerycznego.
movie_num = movie_new.select_dtypes(include= ['number'])
movie_num.head(5)
```

```python
# 3. Wypełnić brakujące wartości w ramce movie_num średnią arytmetyczną wyznaczoną dla każdej kolumny z osobna. 
means = movie_num.describe().loc['mean']

#movie_num = movie_num.fillna(means)
movie_num.fillna(means, inplace=True)
print(means)
movie_num.head(5)
```

```python
# 4. Znaleźć kolumnę w ramce movie_num, której wartości mają największe odchylenie standardowe.
column_std = movie_num.std()
column_with_max_std = column_std.idxmax()
print(f' kolumna z najwyzszym std: {column_with_max_std}\n std dla wszystkich kolumn: \n{column_std}')
```

```python
# 5. Zastąpić kolumny numeryczne w ramce movie_new nowymi kolumnami z ramki movie_num.
movie_new.update(movie_num)
movie_new.head()
```

```python
# 6. Zastąpić wszystkie wartości wybrakowane w ramce movie_new za pomocą pierwszej wartości występującej w kolumnie.
movie_new.fillna(method='ffill', axis=0, inplace=True)
movie_new.head()
```

```python
# 7. Nadać nowe indeksy wierszom w ramce movie_new według następującego schematu: [imie i nazwisko rezysera] - [tytul filmu].
movie_new['new_index'] = movie_new['director_name'] + ' - ' + movie_new['movie_title']
movie_new.set_index('new_index', inplace=True)
movie_new.head()
```

```python
# 8. Zwiększyć wartość kolumny imdb_score każdego filmu w ramce movie_new o 1.
movie_new['imdb_score'] = movie_new['imdb_score'] + 1
movie_new.head()
```

```python
# 9. Zmienić nazwy wszystkich kolumn w ramce movie_new w taki sposób aby zlikwidować wszystkie znaki _ oraz pierwsze litery wszystkich słów w nazwach były wielkie.
movie_new.rename(columns=lambda x: x.replace('_', ' ').title(), inplace=True)
movie_new.head()
```

```python
# 10. Utworzyć ramkę movie_random, w której zostaną wybrane wiersze o losowej ilości w zakresie pozycji [start, end], gdzie start i end będą liczbami pseudolosowymi.

start = np.random.randint(0, len(movie_new))
end = np.random.randint(start, len(movie_new))

# Sample rows within the specified range
movie_random = movie_new.iloc[start:end+1].copy()
print(f'start: {start}, end: {end}')
movie_random.describe()
```

```python
# 11. Zwiększyć wartości każdej kolumny numerycznej w ramce movie_random o wartość pseudolosową (po jednej wartości na kolumnę).
numeric_columns = movie_random.select_dtypes(include=['number'])

def add_random_value(column):
    return column + np.random.normal(8,10)

movie_random[numeric_columns.columns] = numeric_columns.apply(add_random_value)

movie_random.head()

```
